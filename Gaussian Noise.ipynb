```python
import cv2
import os

def add_gaussian_noise(video_path, output_path, mean=0.5, var=0.01):
  """
  Adds Gaussian noise to a video file.

  Args:
      video_path (str): Path to the input video file.
      output_path (str): Path to save the noise-added video file.
      mean (float, optional): Mean value of the Gaussian noise. Defaults to 0.
      var (float, optional): Variance of the Gaussian noise. Defaults to 0.01.
  """

  cap = cv2.VideoCapture(video_path)

  if not cap.isOpened():
    print("Error opening video:", video_path)
    return

  # Get video properties
  width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
  height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
  fps = cap.get(cv2.CAP_PROP_FPS)
  fourcc = cv2.VideoWriter_fourcc(*'XVID')  # Adjust fourcc for your desired output format

  out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

  while True:
    ret, frame = cap.read()

    if not ret:
      break

    # Add Gaussian noise to each frame
    noise = np.random.normal(mean, var, size=frame.shape).astype(np.float32)
    noisy_frame = frame + noise

    # Clip pixel values to [0, 255]
    noisy_frame = np.clip(noisy_frame, 0, 255).astype(np.uint8)

    out.write(noisy_frame)

  cap.release()
  out.release()
  cv2.destroyAllWindows()

# Replace with your FaceForensics++ dataset directory
dataset_dir = "/content/drive/MyDrive/Deepfake Project/FF_Face_only_data"

# Output directory for noise-added videos (create if it doesn't exist)
output_dir = os.path.join(dataset_dir, "noisy_videos")
os.makedirs(output_dir, exist_ok=True)

# Iterate through each video in the dataset and add noise
for filename in os.listdir(dataset_dir):
  if filename.endswith(".mp4") or filename.endswith(".avi"):  # Adjust for your video formats
    video_path = os.path.join(dataset_dir, filename)
    output_path = os.path.join(output_dir, filename)
    add_gaussian_noise(video_path, output_path)

print("Gaussian noise added to videos. Check the 'noisy_videos' directory.")
```
